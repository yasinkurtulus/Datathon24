```python
def turkish_lowercase(text):
    replacements = {
        'İ': 'i',
        'I': 'ı',
        'Ş': 'ş',
        'Ç': 'ç',
        'Ğ': 'ğ',
        'Ö': 'ö',
        'Ü': 'ü',
        'ı': 'ı',
        'ş': 'ş',
        'ç': 'ç',
        'ğ': 'ğ',
        'ö': 'ö',
        'ü': 'ü'
    }
    return ''.join(replacements.get(char, char) for char in text).lower()
```

```python
train["Anne Egitim Durumu"]=train["Anne Egitim Durumu"].apply(turkish_lowercase)
```
