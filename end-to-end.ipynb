```python

def drop_cols(features,df):
    for col in features:
        if col in df.columns:
            df.drop(col,axis=1,inplace=True)
            
```

```python
dropcol=["Burslu ise Burs Yuzdesi","Daha Once Baska Bir Universiteden Mezun Olmus","Lise Adi Diger","Lise Bolum Diger","Burs Aldigi Baska Kurum","Baska Kurumdan Aldigi Burs Miktari",
        "Anne Sektor","Baba Sektor","Uye Oldugunuz Kulubun Ismi","Hangi STK'nin Uyesisiniz?","Uye Oldugunuz Kulubun Ismi","Girisimcilikle Ilgili Deneyiminiz Var Mi?","Girisimcilikle Ilgili Deneyiminizi Aciklayabilir misiniz?","Ingilizce Seviyeniz?","Daha Önceden Mezun Olunduysa, Mezun Olunan Üniversite"]
```

```python
drop_cols(dropcol,train_data)
```

```python
## fiiling numeric value with median
for label,content in train_data.items():
    if pd.api.types.is_numeric_dtype(content) and (content.isna().sum())>0:
         train_data[label]=content.fillna(content.median())
```

```python
## fiiling cat value with null
for label,content in train_data.items():
    if pd.api.types.is_object_dtype(content) and (content.isna().sum())>0:
         train_data[label]=content.fillna("missing")
```

```python

def categorical_columns(df):
    cat_columns=[]
    for col in df.columns:
        if df[col].dtypes=="O" and col!="Burs Aliyor mu?":
            cat_columns.append(col)
    return cat_columns
train_categorical=categorical_columns(train_data)
```
